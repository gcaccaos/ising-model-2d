```python
# Importacao de pacotes
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
from ising2d import *

np.seterr(divide = 'ignore', invalid = 'ignore'); # ignora warnings sobre divisao por zero (no calculo da probabilidade de flipar os spins para T = 0)
```

```python
@jit
def display_spins(spins):
    plt.imshow(spins, cmap = plt.cm.binary)
    plt.axis('off')
    plt.show()
```

```python
@jit
def magnetization_per_site(spins):
    L = spins.shape[0]
    M = spins.sum()
    m = M/(L**2)
    
    return m
```

```python
@jit
def energy_per_site(spins):
    L = spins.shape[0]
    E = 0
    
    for i in range(L):
        for j in range(L):
            DeltaE = 2*spins[i, j]*(J/4*(spins[(i - 1)%L, j] + spins[(i + 1)%L, j] + # dividido por 4 para nao constar o mesmo spin duas vezes
                                       spins[i, (j - 1)%L] + spins[i, (j + 1)%L]) + h)
            E -= DeltaE
    
    e = E/(L**2)
    
    return e
```

```python
def display_ising_sequence(images):
    def _show(step = (0, len(images) - 1)):
        display_spins(images[step])
    return interact(_show)
```

```python
# Inicializacao das variaveis da simulacao
L = 16 # tamanho do array de spins
J = 1 # constante de interacao entre os spins
h = 0 # constante de interacao do spin com o campo magnetico externo

T = 0.0001

Nsteps = 10000
m = np.zeros(Nsteps)
e = np.zeros(Nsteps)

images = [choice([-1, +1], (L, L))]
for k in range(Nsteps):
    images.append(ising_step(images[-1].copy(), T, J, h))
    m[k] = magnetization_per_site(images[k])
    e[k] = energy_per_site(images[k])

display_ising_sequence(images);

plt.figure()
plt.plot(np.abs(m))
plt.xlabel('Step')
plt.ylabel('Absolute magnetization per site')
plt.ylim(-0.1, 1.1)

plt.figure()
plt.plot(e)
plt.xlabel('Step')
plt.ylabel('Energy per site')

plt.show()
```

```python
# Inicializacao das variaveis da simulacao
L = 8 # tamanho do array de spins
J = 1 # constante de interacao entre os spins
h = 0 # constante de interacao do spin com o campo magnetico externo

Tmin = 1e-10 # temperatura proxima de zero
Tmax = 5
DeltaT = 0.5
T = np.arange(Tmin, Tmax + DeltaT, DeltaT)
NT = len(T)

Nsteps = 10000

m = np.zeros(NT)
e = np.zeros(NT)

for k in range(NT):
    thism = np.zeros(Nsteps)
    thise = np.zeros(Nsteps)

    images = [choice([-1, +1], (L, L))]
    for n in range(Nsteps):
        images.append(ising_step(images[-1].copy(), T[k], J, h))
        thism[n] = magnetization_per_site(images[n])
        thise[n] = energy_per_site(images[n])
    
    m[k] = np.mean(np.abs(thism))
    e[k] = np.mean(thise)

plt.figure()
plt.plot(T, m, '-o')
plt.xlabel('Temperature / kB')
plt.ylabel('Absolute magnetization per site')
plt.ylim(-0.1, 1.1)

plt.figure()
plt.plot(T, e, '-o')
plt.xlabel('Temperature / kB')
plt.ylabel('Energy per site')
plt.show()
```
